```python
q = """select a.app_id, sum(case when a.is_recommended = 0 then 1 else 0 end) as not_recommended, sum(a.is_recommended) as is_really_recommended
        from t2 a
        group by a.app_id
        ;
        """
t3 = pysqldf(q)
```
