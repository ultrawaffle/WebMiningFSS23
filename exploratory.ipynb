```python
import pandas as pd
games_table = pd.read_csv('games.csv')
users_table = pd.read_csv('users.csv')
recommendations_table = pd.read_csv('recommendations.csv')
```

```python
display(recommendations_table.head())
```

```python
display(users_table)
```

```python
#Remove columns 'win' 'mac' 'linux' 'discount' 'steam_deck' from table
games_table = games_table.drop(['win', 'mac', 'linux', 'discount', 'steam_deck'], axis=1)
display(games_table)
```

```python
games_table.describe()
```

```python
merged_GR = pd.merge(games_table, recommendations_table, on='app_id')
display(merged_GR)
```

```python
#Remove 'title' 'date_release' 'win' 'mac' 'linux' 'discount' 'steam_deck' 'user_reviews' 'price_final' 'price_original' 'date' from table
merged_GR = merged_GR.drop(['title', 'date_release', 'win', 'mac', 'linux', 'discount', 'steam_deck', 'user_reviews', 'price_final', 'price_original', 'date'], axis=1)
display(merged_GR)
```

```python
# Normalise column attribue 'hours' using min-max normalisation
merged_GR['hours'] = (merged_GR['hours'] - merged_GR['hours'].min()) / (merged_GR['hours'].max() - merged_GR['hours'].min())
display(merged_GR)
```

```python
#import scikit-learn library
import sklearn
from sklearn.preprocessing import LabelEncoder

# Enocde values for column attribute 'is_recommend' using label encoding
label_encoder = LabelEncoder()
merged_GR['is_recommended'] = label_encoder.fit_transform(merged_GR['is_recommended'])
display(merged_GR)
```

```python
# Display range of values for column attribute 'rating'
merged_GR['rating'].unique()
```

```python
# Replace values for column attribute 'rating' with 'Mostly Negative as -1' 'Mixed as 0' 'Positive as 1' 'Mostly Positive as 2' 'Very Positive as 3' 'Overwhelmingly Positive as 4'
merged_GR['rating'] = merged_GR['rating'].replace('Mostly Negative', -1)
merged_GR['rating'] = merged_GR['rating'].replace('Mixed', 0)
merged_GR['rating'] = merged_GR['rating'].replace('Positive', 1)
merged_GR['rating'] = merged_GR['rating'].replace('Mostly Positive', 2)
merged_GR['rating'] = merged_GR['rating'].replace('Very Positive', 3)
merged_GR['rating'] = merged_GR['rating'].replace('Overwhelmingly Positive', 4)
display(merged_GR)
```

```python
# Visualize the correlation between the column attributes 'rating', 'hours' and 'is_recommended'
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
sns.heatmap(merged_GR.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Creating a boxplot to identify outliers for column attribute 'hours'
sns.boxplot(x=merged_GR['hours'])
plt.show()
```
